```python

# Step 2: Create Grid Cells Over Study Area
# Define grid parameters: cell size and study area boundaries (min/max latitude and longitude)
cell_size = 0.1  # Define the grid cell size (adjust based on your study area)
min_lat, max_lat = min(fire['latitude']), max(fire['latitude'])
min_lon, max_lon = min(fire['longitude']), max(fire['longitude'])

# Create grid
grid_lats = np.arange(min_lat, max_lat, cell_size)
grid_lons = np.arange(min_lon, max_lon, cell_size)

# Step 3: Assign Fire Data to Grid Cells and Create Occurrence Matrix
fire_occurrences = np.zeros((len(grid_lats), len(grid_lons)), dtype=int)  # Matrix to store fire occurrence per grid cell

# Loop through fire data and assign to grid cells
for idx, fire_entry in fire.iterrows():
    lat_idx = np.searchsorted(grid_lats, fire_entry['latitude']) - 1
    lon_idx = np.searchsorted(grid_lons, fire_entry['longitude']) - 1

    # Mark presence of fire for corresponding grid cell
    fire_occurrences[lat_idx][lon_idx] += 1  # Increment occurrence count

# Convert the matrix into a DataFrame
grid_cells = []
for i, lat in enumerate(grid_lats):
    for j, lon in enumerate(grid_lons):
        grid_cells.append({'Latitude': lat, 'Longitude': lon, 'Fire_Occurrence': fire_occurrences[i][j]})

fire_occurrence_df = pd.DataFrame(grid_cells)

# Step 4: Output the new dataset representing fire occurrences in grid boxes
fire_occurrence_df.to_csv('fire_occurrence_dataset.csv', index=False)  # Save as CSV or desired format
```

```python

from datetime import datetime


# Convert timestamp column to datetime if it's not already in datetime format
fire['acq_date'] = pd.to_datetime(fire['acq_date'])

# Step 2: Define the range of years (2001 to 2023)
start_year = 2001
end_year = 2023

# Step 3: Create Grid Cells Over Study Area
# Define grid parameters: cell size and study area boundaries (min/max latitude and longitude)
cell_size = 0.1  # Define the grid cell size (adjust based on your study area)
min_lat, max_lat = min(fire['latitude']), max(fire['latitude'])
min_lon, max_lon = min(fire['longitude']), max(fire['longitude'])

# Create grid
grid_lats = np.arange(min_lat, max_lat, cell_size)
grid_lons = np.arange(min_lon, max_lon, cell_size)

# Step 4: Aggregate Fire Occurrences by Yearly and Monthly Frequencies
fire_occurrences_yearly = np.zeros((len(grid_lats), len(grid_lons), end_year - start_year + 1), dtype=int)
fire_occurrences_monthly = np.zeros((len(grid_lats), len(grid_lons), (end_year - start_year + 1) * 12), dtype=int)

# Loop through fire data and assign to grid cells by year and month
for idx, fire_entry in fire.iterrows():
    lat_idx = np.searchsorted(grid_lats, fire_entry['latitude']) - 1
    lon_idx = np.searchsorted(grid_lons, fire_entry['longitude']) - 1
    year_index = fire_entry['acq_date'].year - start_year
    month_index = (fire_entry['acq_date'].year - start_year) * 12 + fire_entry['acq_date'].month - 1

    # Increment occurrence count for yearly and monthly aggregations
    fire_occurrences_yearly[lat_idx][lon_idx][year_index] += 1
    fire_occurrences_monthly[lat_idx][lon_idx][month_index] += 1


# Step 5: Convert the matrices into DataFrames for each frequency (yearly and monthly)
# Code logic to create DataFrames is similar to the previous example, but adjusted for yearly and monthly data
# Convert the matrix into a DataFrame
grid_cells = []
for i, lat in enumerate(grid_lats):
    for j, lon in enumerate(grid_lons):
        grid_cells.append({'Latitude': lat, 'Longitude': lon, 'Fire_Occurrence': fire_occurrences_yearly[i][j]})
        

        fire_occurrence_yearly_df = pd.DataFrame(grid_cells)
# Save DataFrames as CSV files or use them as required
# Example:
fire_occurrences_yearly_df.to_csv('fire_occurrences_yearly.csv', index=False)
fire_occurrences_monthly_df.to_csv('fire_occurrences_monthly.csv', index=False)
```
