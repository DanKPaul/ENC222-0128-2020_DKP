```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
dat = pd.read_csv("CSV/point.csv")
dat
```

```python
dat
```

```python
dat.describe()
```

```python
dat["PARAMETER"].unique()
```

```python
TS = dat[dat.PARAMETER=='TS']
QV2M= dat[dat.PARAMETER=='QV2M']
RH2M = dat[dat.PARAMETER=='RH2M']
WD10M = dat[dat.PARAMETER=='WD10M']
T2M_MAX = dat[dat.PARAMETER=='T2M_MAX']
T2M_MIN = dat[dat.PARAMETER=='T2M_MIN']
PRECTOTCORR = dat[dat.PARAMETER=='PRECTOTCORR']
WS10M_RANGE = dat[dat.PARAMETER=='WS10M_RANGE']
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Surface Temperature Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Surface Temeperature(C)", fontsize=15)
plt.xticks(range(2000, 2023))
plt.plot(TS['YEAR'],TS['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Specific Humidity Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Specific Humidity at 2m(g/kg)", fontsize=15)
plt.xticks(range(2000, 2023))
plt.plot(QV2M['YEAR'],QV2M['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Relative Humidity Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Relative Humidity at 2m(%)", fontsize=15)
plt.xticks(range(2000, 2023))
plt.plot(RH2M['YEAR'],RH2M['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Wind Direction Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Wind Direction at 10m(DEGREES)", fontsize=15)
plt.xticks(range(2000, 2023,+1))
plt.plot(WD10M['YEAR'],WD10M['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Wind Speed Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Wind Speed at 10m range(m/s)", fontsize=15)
plt.xticks(range(2000, 2023,+1))
plt.plot(WS10M_RANGE['YEAR'],WS10M_RANGE['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Maximum Temperature Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Maximum Temperature(C)", fontsize=15)
plt.xticks(range(2000, 2023,+1))
plt.plot(T2M_MAX['YEAR'],T2M_MAX['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Minimum Temperature Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Minimum Temperature(C)", fontsize=15)
plt.xticks(range(2000, 2023,+1))
plt.plot(T2M_MIN['YEAR'],T2M_MIN['ANN'])
```

```python
sbn.set()
plt.figure(figsize=(20,10))
plt.title("Precipitation Time Series", fontsize=20)
plt.xlabel('Years', fontsize=15)
plt.ylabel("Precipitation(mm/day)", fontsize=15)
plt.xticks(range(2000, 2023,+1))
plt.plot(PRECTOTCORR['YEAR'],PRECTOTCORR['ANN'])
```
